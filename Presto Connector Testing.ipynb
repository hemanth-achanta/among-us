```python
# Creating Connection and Initializing Cursor

def presto_connection(host, port, user):
    try:
        conn = prestodb.dbapi.connect(host=host,
                                      port=port,
                                      user=user,
                                      http_scheme="https",
                                      auth=prestodb.auth.BasicAuthentication(user, password),
                                      )
        # isolation_level=transaction.IsolationLevel.REPEATABLE_READ)
    except Exception as e:
        print(str(e))
    return conn


# Function for Retrieving Dataframe using Presto Connection
def df_from_presto(query):
    df = pd.DataFrame()
    try:
        print("Establishing Connection")
        conn = presto_connection(host, port, user)
        conn._http_session.verify = False
        print("Connected")
        cursor = conn.cursor()
        print("Starting Query Execution")
        cursor.execute(query)
        records = cursor.fetchall()
        print("Query Executed")
        column_descriptions = cursor.description
        df = pd.DataFrame(records)
        df.columns = [c[0] for c in column_descriptions]
        # print(df)
        conn.close()
        print("Connection Closed")
    except Exception as e:
        print("Error Message: ", str(e))
    return df
```
